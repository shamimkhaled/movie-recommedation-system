## Importing the dependencies

```python
import numpy as np
import pandas as pd
import difflib # differentiations between these things to that things
from sklearn.feature_extraction.text import TfidfVectorizer # convert the texual data to numerical value feature vectorization
from sklearn.metrics.pairwise import cosine_similarity
```

## Data Collection and Preprocessing

```python
# loading the data from the csv file to apandas dataframe
movies_df = pd.read_csv('movies.csv')
# display the first 5 rows of the dataframe or dataset
movies_df.head()
```

```python
# number of rows and columns in the dataset
movies_df.shape
```

```python
# what are the data types of all the columns that are present
movies_df.info()
```

```python
## Check the missing values
movies_df.isnull().sum()
```

```python
# replacing the null valuess with null string
movies_df['homepage'] = movies_df['homepage'].fillna('')
movies_df['overview'] = movies_df['overview'].fillna('')
movies_df['release_date'] = movies_df['release_date'].fillna('')
#movies_df['release_date'] = movies_df['release_date'].fillna(movies_df['release_date'].median())
movies_df['runtime'] = movies_df['runtime'].fillna('')
```

```python
# choosing the relevant features for movie recommendation
choosen_features = ['genres','keywords','tagline','cast','director']
print(choosen_features)
```

```python
# replacing the null valuess with null string
for features in choosen_features:
  movies_df[features] = movies_df[features].fillna('')
```

```python
# Again Check the missing values
movies_df.isnull().sum()
```

```python
# combining all the 5 selected features
combined_choosen_features = movies_df['genres']+' '+movies_df['keywords']+' '+movies_df['tagline']+' '+movies_df['cast']+' '+movies_df['director']
print(combined_choosen_features)
```

```python
# convert the text data to feature vectors
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_choosen_features)
```

```python
print(feature_vectors)
```

```python
# getting the similarity scores using cosine similarity
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
similarity.shape
```

```python
# creating a list with all the movie names given in dataset
all_movie_titles_list = movies_df['title'].tolist()
print(all_movie_titles_list)
```

```python
# getting the movie name from the user input
movie_name = input(' Enter the movie name : ')
```

```python
# finding the close match for the movie name given by the user iput
find_closing_match = difflib.get_close_matches(movie_name, all_movie_titles_list)
print(find_closing_match)
```

```python
# find closing match for movie name
close_match = find_closing_match[0]
print(close_match)
```

```python
# finding the index of the movie with title given by above user input
index_num_of_movie = movies_df[movies_df.title == close_match]['index'].values[0]
print(index_num_of_movie)
```

```python
# getting a list of similar movies
similarity_score = list(enumerate(similarity[index_num_of_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index
print('Suggested Movies for you : \n')
i = 1
for movies in sorted_similar_movies:
  index = movies[0]
  title_based_on_index = movies_df[movies_df.index==index]['title'].values[0]
  if (i < 20):
    print(i, '.',title_based_on_index)
    i+=1
```

## Movie Recommendation Sytem

```python
movie_name = input(' Enter the movie name : ')

# creating a list with all the movie names given in dataset
all_movie_titles_list = movies_df['title'].tolist()

# finding the close match for the movie name given by the user iput
find_closing_match = difflib.get_close_matches(movie_name, all_movie_titles_list)

# find closing match for movie name
close_match = find_closing_match[0]

# finding the index of the movie with title given by user input
index_num_of_movie = movies_df[movies_df.title == close_match]['index'].values[0]

# getting a list of similar movies
similarity_score = list(enumerate(similarity[index_num_of_movie]))

# sorting the movies based on their similarity score
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

# print the name of similar movies based on the index
print('Suggested Movies for you : \n')
i = 1
for movies in sorted_similar_movies:
  index = movies[0]
  title_based_on_index = movies_df[movies_df.index==index]['title'].values[0]
  if (i < 20):
    print(i, '.',title_based_on_index)
    i+=1
```
